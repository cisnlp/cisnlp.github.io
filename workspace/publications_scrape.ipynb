```python
error = []
for year in ['2020','2019','2018','2017']:
    for pub in all_pubs[year].findChildren('li'):
        url=None
        try:
            authors = pub.contents[0]
            if pub.find('span') is None: 
                title = pub.find('b').text
            else:
                title = pub.find('span').text
#             title = pub.find('span').text
            proceedings = pub.find('i').text
            if pub.find('a') is not None: url = pub.find('a')['href']
            pubs_df = pubs_df.append({'year':year,'authors':authors,'title':title,'proceedings':proceedings,'url':url},ignore_index=True)
        except:
            error.append(pub)
```
